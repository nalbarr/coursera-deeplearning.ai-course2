```python
def a1(t):
    alpha = np.exp(t)
    return alpha
```

```python
def a2(t):
    alpha = (1/(1+2*t))
    return alpha
```

```python
def a3(t):
    alpha = 1/np.sqrt(t)
    return alpha
```

```python
def a4(t):
    alpha = 0.95**t
    return alpha
```

```python
for t in range(100):
    print("a1: " + str(a1(t)) + " a2: " + str(a2(t)) + " a3: " + str(a3(t)) + " a4: " + str(a4(t)))
```
